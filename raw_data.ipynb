## Cria um dataframe copia para come√ßar a limpeza dos dados. O primeiro passo ser√° definir o tipo das colunas que n√£o ser√£o de texto.

```python
df_listings_clean = df_listings.copy()

list_datetime = [ 'last_scraped', 'host_since', 'calendar_updated', 'calendar_last_scraped', 'first_review', 'last_review'] # Colunas do tipo de data e hora
for column in list_datetime:
    # Transforma as colunas para o tipo "datetime" com o formato passado
    df_listings_clean[column] = pd.to_datetime(df_listings_clean[column], format='%Y-%m-%d')

list_int = ['id', 'scrape_id', 'host_id', 'accommodates', 'minimum_nights', 'maximum_nights', 'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights', 'availability_30', 'availability_60', 'availability_90', 'availability_365', 'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms'] #colunas do tipo inteiro
for column in list_int:
    # Transforma as colunas para o tipo inteiro
    df_listings_clean[column] = df_listings_clean[column].astype("int")

list_float = ['host_response_rate', 'host_acceptance_rate', 'host_listings_count', 'host_total_listings_count', 'bedrooms', 'beds', 'price', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'reviews_per_month'] #colunas do tipo float
for column in list_float:
    if column in df_listings_clean.columns:
        # Tratamento para tirar caracteres especiais da coluna (% e $) e salvar no formato float
        if column == 'host_response_rate' or column == 'host_acceptance_rate':
            df_listings_clean[column] = df_listings_clean[column].str.replace('%', '').astype(float) / 100
        elif column == 'price':
            df_listings_clean[column] = df_listings_clean[column].str.replace('$', '').str.replace(',', '').astype(float)
        else:
            df_listings_clean[column] = df_listings_clean[column].astype(float)
```

```python
df_listings_clean.dtypes
```

```python
# Fun√ß√£o pra verificar dados nulos (qual o percentual de dados nulos em cada coluna)
def check_missing(df):
    res_missing = df_listings_clean.isna().sum()
    res_missing = (res_missing/len(df_listings_clean))*100
    return res_missing

check_missing(df_listings_clean).sort_values(ascending=False)
```

#### Dropa colunas com muitos nulos

```python
df_listings_clean = df_listings_clean.drop(columns=['calendar_updated', 'license', 'neighbourhood_group_cleansed', 'bathrooms'])
```

```python
# check nulos
check_missing(df_listings_clean).sort_values(ascending=False)
```

#### Elimina dados duplicados

```python
print("N. de linhas antes de remover duplicadas:", len(df_listings_clean))
df_listings_clean = df_listings_clean.drop_duplicates()
print("N. de linhas depois de remover duplicadas:", len(df_listings_clean))
```

#### Elimina colunas com dados constante

```python
list_constant = [col for col in df_listings_clean.columns if df_listings_clean[col].nunique() == 1]
list_constant # ['scrape_id'] visto que os dados s√£o apenas do Rio de Janeiro
```

```python
print("N. de colunas antes de remover colunas constantes:", len(df_listings_clean.columns))
df_listings_clean = df_listings_clean.drop(list_constant, axis=1)
print("N. de colunas depois de remover colunas constantes:", len(df_listings_clean.columns))
```

## Gerando novo resultado

```python
# profile = ProfileReport(df_listings_clean, title="Pandas Profiling Report")

# profile.to_file("new_results_listings.html")
```

```python
df_listings_clean.to_csv('listings_clean.csv')
```
